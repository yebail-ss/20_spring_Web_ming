## 数据科学家使用xpath的角度
![](https://qxf2.com/blog/wp-content/uploads/2015/12/Table.png)
* 不挑greedy的策略
    * 求全求不漏
    * 可能有垃圾
    * 使用 // （后代descendant） 而不用  / 子女（children）
    * 使用 **任意**元素/标签  而不用  **指定**元素/标签
* 挑剔ungreedy的策略
    * 求准求数据整齐
    * 可能有漏数据
    * 使用  / 子女（children） 而不用  // （后代descendant）
    * 使用 **指定**元素/标签  而不用  **任意**元素/标签 

### 不挑greedy

````python
# B-B-1 使用 xpath 
r.html.xpath('//a')  # html页面内所有的a标签都爬取过来（不选择）
````

````python
# B-B-2 使用 xpath 限定 取特定属性
# 注意和 B1 的内容相比, 是不是少了一些? 
# 没有特定属性title就不会被选到
解析后.xpath('//a/@title')  # 爬取a标签下面的所有title的值
````

````python
# B-B-3-1 使用 xpath  理解不同可以有不同的表达，但表示效果一可以一样
# greedy 不挑元素/标签  只挑有特定属性title的所有元素
print(r.html.xpath('//*/@title'))
# greedy 不挑位置，只挑有特定属性title的所有元素
print(r.html.xpath('//@title'))
````

### 挑剔ungreedy

````python
# B-B-4 使用 xpath  # ungreedy 更精準
r.html.xpath('//div[@class="news_title"]/a/@title')
````

````python
# B-B-5 使用 xpath  # ungreedy 最精准
# xpath 完全没有用 // 也没有用 * 
# 从路径上理解，相当于绝对路径的概念，虽然可以找到我们要的元素，但少了一个环节就会出错
# 检查难度也更大，易读性也不强，不是很推荐这种写法
print(r.html.xpath('body/div[@class="lin-content"]/div[@class="lin-neiye clearfix"]/div[@class="search_list_right"]/div[@class="ny_content"]/ul[@class="list-ul"]/li/div[@class="news_title"]/a'))

# 换种写法，代码会简洁很多
print(r.html.xpath('//div[@class="news_title"]/a'))

````

# 本周小结内容
## 一、节点 
- 在 XPath 中，有七种类型的节点：元素、属性、文本、命名空间、处理指令、注释以及文档（根）节点。XML文档是被作为节点树来对待的。树的根被称为文档节点或者根节点。
- html以元素/标签构成
- 值和文本不单独存在，必需依附元素/标签
- 获取标签tags丶属性attributes丶值values
- 掌握Chrome Inspector 多种颜色区分

     * 元素/标签 elements/tags  紫色
     * 属性attributes 橙色
     * 值values 蓝色
     * 文本 黑色
     * HTML注解 灰色

```
<book>
    <author class="classical">J K. Rowling</author>
    <year>2005</year>
    <price>29.99</price>
</book> 
```
- **<book> 是文档节点**
- **class="classical"是属性节点**
- **<year>2005</year>是元素节点**

## 二、节点之间的关系有：
**父节点、子节点、同胞、前辈、后代**
```
<li class="flexbox">
<a class="flex-2 job-name" href="/register/?imscid=R000010827">
<span class="name-text">pandas招聘专场进行时</span>
<img class="fire" src="https://image0.lietou-static.com/img/58ca5a0d70321f8a8b37f94f05a.png" alt="">
</a>
<span class="text-warning flex-1">10万起</span>
</li>
```
- 以```<img class="fire">```为基础，```<span class="name-text">```是它的同胞，```<a class="flex-2 job-name">```是他的父节点，```<li class="flexbox">```是它的先辈
- 以```<li class="flexbox">```为基础，```<a class="flex-2 job-name">``` 是它的子节点，```<img class="fire">```是它的后代

## 三、选取节点相关知识
- "nodename" 表示选取此节点的所有子节点
- "/"从根节点选取
- "//"从匹配选择的当前节点选择文档中的节点，而不考虑位置
- "."选取当前的节点
- ".."选取当前节点的父节点
- "@"选取属性
- 如果 XPath 的开头是一个斜线（/）代表这是绝对路径。如果开头是两个斜线（//）表示文件中所有符合模式的元素都会被选出来，即使是处于树中不同的层级也会被选出来。
### 具体例子
- $x('li') 选取li元素所有子节点
- $x('/li')选取根元素li
- $x('/li/a')选取属于li的子元素的所有a元素
- $x('//li')选取所有li子元素，不考虑位置
- $x('li//a')选取属于li元素的后代所有a元素而不管他们位于li元素之下的哪个位置
- $x('//@class') 选取名为class的所有属性
